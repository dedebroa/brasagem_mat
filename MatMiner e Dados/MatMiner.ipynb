### INTRODUÇÃO AO MATMINER

<div style=' text-align: justify; text-justify: inter-word;'>
  De acordo com seus próprios criadores, o 'matminer' é uma biblioteca para a mineração de dados sobre inúmeros materiais diferentes. la possui diversos recursos que possibilitam, por exemplo, que você explore dados de datasets desenvolvidos com objetivos específicos; que você  crie seu próprio dataset, que manipule features a fim de transformá-las em dados numéricos. Muitas ações são possíveis e recomendamos ler o artigo "Matminer: An open source toolkit for materials data mining" escrito por Ward, L., Dunn, A., Faghaninia, A., Zimmermann, N. E. R., et al e publicado em Comput. Mater. Sci. 152, 60-69 (2018). Tal artigo também pode ser acessado no seguinte link: <mark> https://www.scholars.northwestern.edu/en/publications/matminer-an-open-source-toolkit-for-materials-data-mining </mark>.
    </div>
