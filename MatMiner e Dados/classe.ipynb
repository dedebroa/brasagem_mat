# Criando a classe para o banco de dados

```python
# imports
from matminer.datasets import load_dataset
import pandas as pd
from matminer.featurizers.conversions import StrToComposition
```

```python
class dataset:
    def __init__(self, df, existente=False):
        if not existente:
            self.df = load_dataset(df)
        else:
            self.df = df

    def __add__(self, other):
        novo_df = pd.merge(self.df, other.df, how="inner", on="composition")
        novo = dataset(novo_df, True)
        return novo

    def cria_composicao(self):
        stc = StrToComposition()
        self.df = stc.featurize_dataframe(self.df, "formula", pbar=False)
```
